```python
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import neural_net
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species', palette='Paired', markers=['o', 'v', '^'], size=2.5)
```

```python
X = iris.values[:, :4]
y = iris.values[:, 4]

X = X.astype(float)

lb = preprocessing.LabelBinarizer()
Y = lb.fit_transform(y)

data = [(x.reshape((-1, 1)), y.reshape((-1, 1))) for x, y in zip(X, Y)]

np.random.shuffle(data)

train_test_split = 0.7
split_idx = int(train_test_split*len(data))
training_data, test_data = data[:split_idx], data[split_idx:]
```

```python
model = neural_net.Model([4, 8, 3])
model.fit(training_data, 200, 4, 0.1, test_data)
```
